# 249. Group Shifted Strings

Perform the following shift operations on a string:

Right shift: Replace every letter with the successive letter of the English alphabet, where 'z' is replaced by 'a'. For example, "abc" can be right-shifted to "bcd" or "xyz" can be right-shifted to "yza".
Left shift: Replace every letter with the preceding letter of the English alphabet, where 'a' is replaced by 'z'. For example, "bcd" can be left-shifted to "abc" or "yza" can be left-shifted to "xyz".
We can keep shifting the string in both directions to form an endless shifting sequence.

For example, shift "abc" to form the sequence: ... <-> "abc" <-> "bcd" <-> ... <-> "xyz" <-> "yza" <-> .... <-> "zab" <-> "abc" <-> ...
You are given an array of strings strings, group together all strings[i] that belong to the same shifting sequence. You may return the answer in any order.

 

Example 1:

Input: strings = ["abc","bcd","acef","xyz","az","ba","a","z"]

Output: [["acef"],["a","z"],["abc","bcd","xyz"],["az","ba"]]

Example 2:

Input: strings = ["a"]

Output: [["a"]]

 

Constraints:

1 <= strings.length <= 200
1 <= strings[i].length <= 50
strings[i] consists of lowercase English letters.

````python
from collections import defaultdict
def groupStrings(strings):
    seen = defaultdict(list)
    for st in strings:
        key = []
        if len(st) > 1:
            
            for i in range(len(st)-1):
                key.append((ord(st[i+1])- ord(st[i]))%26)
            seen[tuple(key)].append(st)
        else:
            seen[()].append(st)

    return list(seen.values())
````

````python
strings = ["abc","bcd","ace","aec","zab","acef","xyz","az","ba","a","z"]
groupStrings(strings)
````

````python
from collections import defaultdict

def groupStrings(strings):
    seen = defaultdict(list)
    
    for st in strings:
        if len(st) == 1:
            # single char → no diffs
            key = ()
        else:
            # ordered tuple of diffs (mod 26)
            key = tuple((ord(st[i+1]) - ord(st[i])) % 26 for i in range(len(st)-1))
        
        seen[key].append(st)
    
    return list(seen.values())
````

````python
strings = ["abc","bcd","ace","aec","zab","acef","xyz","az","ba","a","z"]
groupStrings(strings)
````

````python
res = groupStrings(["abc","bcd","xyz"])
print("Result 1:", res)
assert any(set(g)=={"abc","bcd","xyz"} for g in res)

res = groupStrings(["a","z"])
print("Result 2:", res)
assert any(set(g)=={"a","z"} for g in res)

res = groupStrings([])
print("Result 3:", res)
assert res == []
````

# Python/Kubernetes questions

1. **When would you choose `asyncio` over threads? One example.**
   When I have *many concurrent I/O-bound* tasks where I control the stack and can use non-blocking APIs—`asyncio` gives me scalable concurrency in a single thread with low overhead.
   **Example:** a service that must open thousands of HTTP/WebSocket connections to fetch/stream data concurrently using `asyncio.gather`.

   **Note:** A global interpreter lock (GIL) is a mechanism used in computer-language interpreters to synchronize the execution of threads so that only one native thread (per process) can execute basic operations (such as memory allocation and reference counting) at a time.[1] As a general rule, an interpreter that uses GIL will see only one thread to execute at a time, even if it runs on a multi-core processor, although some implementations provide for CPU intensive code to release the GIL, allowing multiple threads to use multiple cores. Some popular interpreters that have a GIL are CPython and Ruby MRI.

2. **`deepcopy` vs `copy` with a nested list.**
   `copy.copy` is *shallow*: it copies the outer container but the inner objects are still shared. `copy.deepcopy` recursively copies everything.
   Example:

```python
import copy
a = [[1], [2]]
b = copy.copy(a)       # shallow
c = copy.deepcopy(a)   # deep
a[0].append(99)
# b -> [[1, 99], [2]]  (shared inner list changed)
# c -> [[1], [2]]      (independent)
```

3. **What problems does the GIL cause, and when does it not matter?**
   The GIL prevents true parallel execution of Python bytecode, so *CPU-bound* threads don’t scale across cores; it can also cause contention/latency. It *doesn’t matter* for I/O-bound workloads (threads spend most time blocked), for multi-process parallelism, or for C/NumPy extensions that release the GIL during heavy work.

4. **Liveness vs Readiness probes (Kubernetes).**
   **Liveness** answers “should the container be restarted?”—fail → kubelet restarts the container. **Readiness** answers “can this pod receive traffic?”—fail → pod is removed from Service endpoints but not restarted. (Startup probe can gate liveness until app is warm.)

5. **Deployment vs StatefulSet; when to pick each.**
   **Deployment**: stateless, interchangeable replicas, rolling updates, scale out web/API tiers.
   **StatefulSet**: stable pod identity/hostnames, ordered (de)ploy, *persistent* volumes per replica—use for databases, Kafka/ZooKeeper, or anything needing sticky identity/storage.

6. **Safely share data between processes in Python.**
   Use **`multiprocessing`** primitives: `Queue`/`Pipe` for message passing, `Manager().dict/list` for shared mutable state (with the right synchronization), or `shared_memory`/`Value`/`Array` for fixed-size buffers; guard critical sections with `Lock`/`RLock`. Prefer message passing (immutable or pickleable payloads) to avoid races.

7. **Clean rollback of a bad Deployment rollout (Kubernetes).**

8. **Make a function cache-friendly for `lru_cache` (small example).**
   Keep it *pure* (no side effects), make arguments *hashable/normalized*, and return deterministic results.

```python
from functools import lru_cache

def expensive_sum(nums):               # accepts list (unhashable)
    return _expensive_sum(tuple(nums)) # normalize

@lru_cache(maxsize=1024)
def _expensive_sum(nums_tuple):
    # pretend this hits disk/network
    return sum(nums_tuple)

# Calls with the same contents hit the cache:
expensive_sum([1,2,3]); expensive_sum([1,2,3])
```

9. **`pip install -e .` vs `pip install .`**
   `pip install -e .` (“editable”) installs a lightweight link to your source tree, so code changes are reflected immediately—ideal for development. `pip install .` builds a wheel and copies artifacts into site-packages; changes to the source won’t apply until you reinstall.

10. **Pod in `CrashLoopBackOff`: first 2–3 commands and why.**

