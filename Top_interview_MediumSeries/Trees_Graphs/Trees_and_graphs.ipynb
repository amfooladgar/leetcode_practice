````python
#Ali's solution using reiterative: complexity is O(N) for both time and space
def inorderTraversal(root):
    stack, res = [], []
    while stack or root:
        while root:
            stack.append(root)
            root = root.left
        root = stack.pop()
        res.append(root.val)
        root = root.right
    return res
````

````python
# Leedcode Solution 3: Morris Traversal and Threaded Binary Tree
# making all the left childs as  parents of root and right childs until there is no left child is left. the result is the in-order traversal
def inorderTraversal(root):
    res = []
    curr = root
    
    while curr is not None:
        if curr.left is None:
            res.append(curr.val)
            curr = curr.right # move to the next right node
        else:
            pre = curr.left
            while pre.right is not None and pre.right != curr:  # find rightmost
                pre = pre.right
            
            if pre.right is None: #establish a link back to current node
                pre.right = curr
                curr = curr.left
            else: #restore the tree structure
                pre.right = None
                res.append(curr.val)
                curr = curr.right
    return res
````

## Explaining the Morris Traversal algorithm: 

ðŸ” Loop: while curr is not None
You iterate through the tree starting from the root (curr).

ðŸ“Œ Case 1: if curr.left is None
This means there's no left subtree, so:

You visit curr (add its value to res)

Then move to its right child.

This handles nodes like leaf nodes or when you've already visited left subtrees.

ðŸŒ± Case 2: else:
When curr has a left child, you need to visit that left subtree first. So:

ðŸ”Ž Step 1: pre = curr.left
You go to the left child.

ðŸ” Step 2: while pre.right is not None and pre.right != curr
You walk to the rightmost node (pre) of the left subtree (i.e., the in-order predecessor of curr).

ðŸ”— Step 3: if pre.right is None â€” First Visit
You temporarily link pre.right = curr so that after the left subtree is done, you can come back to curr without using a stack. Then:

Move curr = curr.left to process the left subtree.

ðŸ§¹ Step 4: else: â€” Second Visit (Thread Exists)
You're coming back from the left subtree via the thread (pre.right == curr). So:

Remove the temporary link: pre.right = None âž âœ… restore the tree

Visit curr: res.append(curr.val)

Move to right: curr = curr.right

âœ… Why It Works
You simulate the call stack by temporarily threading back to the parent (curr) from its in-order predecessor.

You restore the original structure by removing the thread (pre.right = None) when you're done with that branch.
