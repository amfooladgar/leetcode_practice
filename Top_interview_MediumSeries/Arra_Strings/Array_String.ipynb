You must do it in place.
```
Input: matrix = [[1,1,1],[1,0,1],[1,1,1]]
Output: [[1,0,1],[0,0,0],[1,0,1]]
```

```
Input: matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
Output: [[0,0,0,0],[0,4,5,0],[0,3,1,0]]
```

- n == matrix[0].length
- 1 <= m, n <= 200
- -231 <= matrix[i][j] <= 231 - 1
 

Follow up:

- Could you devise a constant space solution?

Hint #1

````python
#Ali's solution: using two Hashsets to save the index of zeros
def setZeroes(matrix):
    zRowIdx = set()
    zColIdx = set()
    m = len(matrix)
    n = len(matrix[0])
    for i in range(m):
        
        for j in range(n):
            if matrix[i][j]==0:
                zRowIdx.add(i)
                zColIdx.add(j)

    for i in range(m):
        if i in zRowIdx:
            matrix[i][:] = [0]*n
        for j in range(n):
            if j in zColIdx:
                matrix[i][j] = 0 
    
````

# Group Anagrams

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

 

Example 1:
```
Input: strs = ["eat","tea","tan","ate","nat","bat"]
Output: [["bat"],["nat","tan"],["ate","eat","tea"]]
```
Explanation:

There is no string in strs that can be rearranged to form "bat".
The strings "nat" and "tan" are anagrams as they can be rearranged to form each other.

The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.

Example 2:
```
Input: strs = [""]
Output: [[""]]
```
Example 3:
```
Input: strs = ["a"]
Output: [["a"]]
```
 

Constraints:

- 1 <= strs.length <= 104
- 0 <= strs[i].length <= 100
- strs[i] consists of lowercase English letters.

````python
#Ali's solution: O(NKlogK): N number of strs and K the max average length of strings
def groupAnagrams(strs):
    seen = {}

    for st in strs:
        key = ''.join(sorted(st))
        if key not in seen:
            seen[key]=[st]
        else:
            seen[key]+=[st]


            
    return list(seen.values())
````

### ChatGPT optimization using Manacher’s algorithm (O(n)): Overall: time: O(N2) and space O(1)
1. Avoid slicing `s[left+1:right]` on every expansion — just return indices, and do slicing once at the end.
2. Optimize early stopping condition: `if len(s) - i <= len(res) // 2:` can be more precise and readable.
3. Avoid unnecessary string comparisons — compare lengths of indices instead of full substrings.
4. Cleaner structure — to make the logic more elegant.

````python
# ChatGPT optimization using Manacher’s algorithm (O(n)): Overall: time: O(N2) and space O(1) if not passing the s to the inner function

def longestPalindrome(s):

    def expandAroundCenter(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return left + 1, right  # return indices instead of the substring
    
    n = len(s)
    if n < 2:
        return s

    start = 0
    max_len = 1

    for i in range(n):
        # Early break optimization
        if (n - i) * 2 <= max_len:
            break

        l1, r1 = expandAroundCenter(i, i)      # Odd-length
        l2, r2 = expandAroundCenter(i, i + 1)  # Even-length

        if r1 - l1 > max_len:
            start = l1
            max_len = r1 - l1
        if r2 - l2 > max_len:
            start = l2
            max_len = r2 - l2

    return s[start:start + max_len]
````

````python
# Leetcode Solution: Expand from Centers
# mostly the same as ChatGPT solution on expanding around center
# Complexity: O(N2) time: O(N) for checking on expandP function and and there are 2N-1 centers. It has O(1) space complexity.

def longestPalindrome(s):
    def expandP(i,j): # returning the lenght of the longest substrign after expand
        left = i 
        right = j

        while left >=0 and right < len(s) and s[left] == s[right]:
            left-=1
            right+=1
        return right - left - 1
    
    ans = [0,0]
    for i in range(len(s)):
        odd_length = expandP(i,i)
        if odd_length > ans[1]-ans[0]+1:
            dist = odd_length//2
            ans = [i-dist,i+dist]

        even_length = expandP(i,i+1)
        if even_length > ans[1]-ans[0] +1:
            dist = (even_length//2) -1
            ans = [i-dist,i+dist+1]
    i , j = ans
    return s[i:j+1]
````

````python
# Leetcode solutino4: Manacher's Algorithm: Complexity: O(N) for both time and space
# Best solution for this problem! But advanced and beyond the scope of the interview questions

def longest_palindrome_manacher(s: str) -> str:
    """
    Manacher’s algorithm – O(n) time / O(n) space
    Returns the longest palindromic substring of s.
    """

    if not s:
        return ""

    # 1) Pre-process: add '#' between every char and sentinels at ends
    T = "^#" + "#".join(s) + "#$"
    n = len(T)

    P = [0] * n          # radius array
    center = radius = 0   # rightmost palindrome so far
    max_len = 0
    max_center = 0

    # 2) Main loop
    for i in range(1, n - 1):                # skip ^ and $
        mirror = 2 * center - i              # symmetric position w.r.t. current center

        # 2a. Initialize P[i] using mirror info if inside current rigradiusht boundary
        if i < radius:
            P[i] = min(radius - i, P[mirror])

        # 2b. Attempt to expand palindrome centred at i
        while T[i + P[i] + 1] == T[i - P[i] - 1]:
            P[i] += 1

        # 2c. If we expanded past 'radius', update center/radius
        if i + P[i] > radius:
            center, radius = i, i + P[i]

        # 2d. Track the longest palindrome so far
        if P[i] > max_len:
            max_len, max_center = P[i], i

    # 3) Convert back to original string indices
    start = (max_center - max_len) // 2
    return s[start: start + max_len]
````

# Increasing Triplet Subsequence

Given an integer array nums, return true if there exists a triple of indices (i, j, k) such that i < j < k and nums[i] < nums[j] < nums[k]. If no such indices exists, return false.

 

Example 1:
```
Input: nums = [1,2,3,4,5]
Output: true
Explanation: Any triplet where i < j < k is valid.
```
Example 2:
```
Input: nums = [5,4,3,2,1]
Output: false
Explanation: No triplet exists.
```
Example 3:
```
Input: nums = [2,1,5,0,4,6]
Output: true
Explanation: The triplet (3, 4, 5) is valid because nums[3] == 0 < nums[4] == 4 < nums[5] == 6.
``` 

Constraints:

- 1 <= nums.length <= 5 * 105
- -231 <= nums[i] <= 231 - 1
 

**Follow up**: Could you implement a solution that runs in O(n) time complexity and O(1) space complexity?

````python
def increasingTriplet(nums):
    n = len(nums)
    if n<3:
        return False
    first_num = float('inf')
    second_num = float('inf')
    for num in nums:
        if num<=first_num:
            first_num = num
        elif num<=second_num:
            second_num = num
        else:
            return True
    return False 
````

# Missing Ranges

You are given an inclusive range `[lower, upper]` and a sorted unique integer array `nums`, where all elements are within the inclusive range. 

A number `x` is considered missing if `x` is in the range `[lower, upper]` and `x` is not in `nums`. 

Return the shortest sorted list of ranges that exactly covers all the missing numbers. That is, no element of nums is included in any of the ranges, and each missing number is covered by one of the ranges.  

Example 1:
```
Input: nums = [0,1,3,50,75], lower = 0, upper = 99
Output: [[2,2],[4,49],[51,74],[76,99]]
Explanation: The ranges are:[2,2][4,49][51,74][76,99]
```

Example 2:
```
Input: nums = [-1], lower = -1, upper = -1O
utput: []
Explanation: There are no missing ranges since there are no missing numbers.
```

Constraints:
- -109 <= lower <= upper <= 109
- 0 <= nums.length <= 100
- lower <= nums[i] <= upper
- All the values of nums are unique.


````python
def findMissingRanges(nums, lower, upper):
    res = []
    prev = lower - 1
    nums.append(upper + 1)  # Sentinel to simplify logic

    for num in nums:
        if num - prev > 1:
            start = prev + 1
            end = num - 1
            res.append([start, end])
        prev = num

    return res
````

````python
nums, lower, upper = [7,8,11,50,75], 6, 99
findMissingRanges(nums, lower, upper)
````

- Now, with the two helper functions, you can start with "1" and call the two functions alternatively n-1 times. The answer is the last integer you will obtain.

````python
def helper_freq(st):
    count = 1
    ans = []
    i = 0
    while i+1 <len(st):
        
        if st[i] == st[i+1]:
            count+=1
        else:
            ans.append([int(st[i]),count])
            count = 1
            
        i+=1
    ans.append([int(st[i]),count])
    return ans

````

````python
# ChatGPT's Optimization of Ali's code
# 1- Removing the unnecessary use of nested lists ([int(st[i]), count]) and reversed(pair)—you can directly build the string.
# 2- Avoiding extra space usage in helper_freq and helper_freq_to_st by combining them into a single step.
# 3- Using string concatenation efficiently.

def countAndSay(n):
    s = "1"
    for _ in range(n - 1):
        result = []
        i = 0
        while i < len(s):
            count = 1
            while i + 1 < len(s) and s[i] == s[i + 1]:
                i += 1
                count += 1
            result.append(str(count) + s[i])
            i += 1
        s = ''.join(result)
    return s
````
