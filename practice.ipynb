# Longest common subsequence


```python
from functool import lru_cache
def longestCommonSubsequence(text1, text2):
    @lru_cache(None)
    def dp(i,j):
        if i==t1 or j==t2:
            return 0
        return 1 + dp(i+1,j+1) if text1[i] == text2[j] else max(dp(i+1,j), dp(i,j+1))

    t1, t2 = len(text1), len(text2)
    return dp(0,0)
```

```python
text1, text2 = "abc", "def"
longestCommonSubsequence(text1, text2)
```

# Maximal Square

```python
from functools import lru_cache
def maximalSquare(matrix):
    @lru_cache(None)
    def dp(i,j):
        
        if i>=m or j>=n:
            return 0
        
        res = 0
        if matrix[i][j]=='1':
            res = 1 + min(dp(i+1,j), dp(i,j+1), dp(i+1,j+1))
        else:
            res = 0 
        return res

    m, n = len(matrix), len(matrix[0])
    max_side = 0
    for i in range(m):
        for j in range(n):
            max_side = max(max_side, dp(i,j))
    
    return max_side ** 2
```

# Minimum Difficulty of a Job Schedule

```python
# practice top-down solution
def minDifficulty(jobDifficulty, d):
    n = len(jobDifficulty)
    if n<d:
        return -1
    hardest_rem = [0]*n #to store the max jobdiff for each index k starting from job k onward, to be used for the remaining jobs of the last day
    cur = 0
    for k in range(n-1,-1,-1):
        cur = max(cur,jobDifficulty[k])
        hardest_rem[k] = cur

    def dp(i,day):
        if day == d:
            return hardest_rem[i]

        ans = float('inf')
        curr_hardest = 0
        for j in range(i,n-(d-day)):
            curr_hardest = max(curr_hardest, jobDifficulty[j])
            ans = min(ans, curr_hardest+dp(j+1,day+1))

        return ans 

    return dp(0,1)

```

```python
jobDifficulty, d = [6,5,4,3,2,1], 2
minDifficulty(jobDifficulty, d)
```

# Search 2D Matrix II

```python
# using the trick of starting from top-right side of the matrix: complexity is O(m+n) for time and O(1) space
def searchMatrix(matrix, target):
    if not matrix:
        return False

    m = len(matrix)
    n = len(matrix[0])

    i = 0
    j = n-1
    while i<m and j>=0:
        if matrix[i][j]==target:
            return True
        elif matrix[i][j]<target:
            i+=1
        elif matrix[i][j]>target:
            j-=1
    return False
```

```python
matrix, target = [[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]], 5
searchMatrix(matrix, target)
```

# Coin Change

```python
def coinChange(coins, amount):
    n = len(coins)
    def dp(rem): #function with state rem
        # base case:
        if rem<0:
            return -1
        if rem == 0:
            return 0
        
        # recursion:
        min_res = float('inf')
        for coin in coins:
            res = dp(rem-coin)
            if res != -1:
                min_res = min(res+1,min_res)
        return min_res if min_res != float('inf') else -1
    
    return dp(amount)
```

```python
coins, amount = [1,2,5], 11
coinChange(coins, amount)
```

# 543. Diameter of Binary Tree

```python
def diameterOfBinaryTree(root):
    diameter = 0
    def dfs(node): #returns the max height from node to deepest node
        if not node:
            return 0
        nonlocal diameter
        left = dfs(node.left)
        right = dfs(node.right)
        diameter = max(diameter, left+right)
        return max(left,right)+1
    dfs(root)
    return diameter
        
        
```

```python
# practice graph problem: Evaluate division problem
def calcEquation(equations, values, queries):
    group_Id_weight = {}  # nodeId: (groupId, weight)
    def find(nodeId):
        if nodeId not in group_Id_weight:
            group_Id_weight[nodeId] = (nodeId, 1)
        
        grouId, nodeWeight = group_Id_weight[nodeId]

        if nodeId != grouId:
            new_gr_Id, weight = find(grouId) 
            group_Id_weight[nodeId] = (new_gr_Id, weight*nodeWeight)
        return group_Id_weight[nodeId]
    
    def union(dividend, divisor, val):
        div_id, div_w = find(dividend)
        divsor_id, divisor_w = find(divisor)

        if div_id != divsor_id:
            group_Id_weight[div_id] = (divsor_id, divisor_w*val/div_w)
    
    for (dividend, divisor), val in zip(equations, values):
        union(dividend, divisor, val)
    res = []
    for (q1, q2) in queries:
        if q1 not in group_Id_weight or q2 not in group_Id_weight:
           res.append(-1.0) 
        else:

            q1_id, q1_w = find(q1)
            q2_id, q2_w = find(q2)

            if q1_id != q2_id:
                res.append(-1)
            else:
                res.append(q1_w/q2_w)
    return res
            
```

```python
equations, values, queries = [["a","b"],["b","c"]], [2.0,3.0], [["a","c"],["b","a"],["a","e"],["a","a"],["x","x"]]
calcEquation(equations, values, queries)
```

```python
# Generate Parentheses problem
def generateParenthesis(n):
    ans = []
    def backtrack(curr, left_cnt, right_cnt):
        if len(curr)==2*n:
            ans.append(''.join(curr[:]))
            return
        if left_cnt<n:
            curr.append("(")
            backtrack(curr,left_cnt+1,right_cnt)
            curr.pop()
        if right_cnt<left_cnt:
            curr.append(")")
            backtrack(curr,left_cnt,right_cnt+1)
            curr.pop()
    backtrack([],0,0) 
    return ans
```

# Longest Subarray of 1's After Deleting One Element

```python
# solving with two pointers in one pass and a window with at most one zero a
def longestSubarray(nums):
    left, n = 0,len(nums)
    zero_count = 0
    best = 0
    for right in range(n):
        if nums[right] == 0:
            zero_count+=1

        while zero_count>1:
            if nums[left] ==0:    
                zero_count-=1
            left+=1

        best = max(best, right-left+1)
        print(best)
    
    return best - 1
```

# Paint House II - practice

```python
def minCostII(costs):

    m = len(costs)
    n = len(costs[0])
    
    if not costs:
        return 0
    prev_min_color, prev_second_min_cost, prev_min_cost = None, None, float('inf')
    for color in range(n):
        curr_cost = costs[0][color]
        if prev_min_cost == None or curr_cost<prev_min_cost:
            prev_second_min_cost = prev_min_cost
            prev_min_color = color
            prev_min_cost = curr_cost
        elif prev_second_min_cost == None or curr_cost<prev_second_min_cost:
            prev_second_min_cost = curr_cost
    
    for house in range(1,m):
        min_color= min_cost= second_min_cost = None
        for color in range(n):
            cost = costs[house][color]
            if color == prev_min_color:
                cost+=prev_second_min_cost
            else:
                cost+=prev_min_cost
            
            if min_cost == None or cost<min_cost:
                second_min_cost = min_cost
                min_cost = cost
                min_color = color
            elif second_min_cost == None or cost<second_min_cost:
                second_min_cost = cost
        
        prev_min_color = min_color
        prev_min_cost = min_cost
        prev_second_min_cost = second_min_cost

    return prev_min_cost


```

```python
costs = [[1,5,3],[2,9,4]]
minCostII(costs)
```
