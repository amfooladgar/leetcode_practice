# Longest common subsequence


```python
from functool import lru_cache
def longestCommonSubsequence(text1, text2):
    @lru_cache(None)
    def dp(i,j):
        if i==t1 or j==t2:
            return 0
        return 1 + dp(i+1,j+1) if text1[i] == text2[j] else max(dp(i+1,j), dp(i,j+1))

    t1, t2 = len(text1), len(text2)
    return dp(0,0)
```

```python
from functools import lru_cache
def maximalSquare(matrix):
    @lru_cache(None)
    def dp(i,j):
        
        if i>=m or j>=n:
            return 0
        
        res = 0
        if matrix[i][j]=='1':
            res = 1 + min(dp(i+1,j), dp(i,j+1), dp(i+1,j+1))
        else:
            res = 0 
        return res

    m, n = len(matrix), len(matrix[0])
    max_side = 0
    for i in range(m):
        for j in range(n):
            max_side = max(max_side, dp(i,j))
    
    return max_side ** 2
```

# Minimum Difficulty of a Job Schedule

```python
# practice top-down solution
def minDifficulty(jobDifficulty, d):
    n = len(jobDifficulty)
    if n<d:
        return -1
    hardest_rem = [0]*n #to store the max jobdiff for each index k starting from job k onward, to be used for the remaining jobs of the last day
    cur = 0
    for k in range(n-1,-1,-1):
        cur = max(cur,jobDifficulty[k])
        hardest_rem[k] = cur

    def dp(i,day):
        if day == d:
            return hardest_rem[i]

        ans = float('inf')
        curr_hardest = 0
        for j in range(i,n-(d-day)):
            curr_hardest = max(curr_hardest, jobDifficulty[j])
            ans = min(ans, curr_hardest+dp(j+1,day+1))

        return ans 

    return dp(0,1)

```

# Search 2D Matrix II

```python
# using the trick of starting from top-right side of the matrix: complexity is O(m+n) for time and O(1) space
def searchMatrix(matrix, target):
    if not matrix:
        return False

    m = len(matrix)
    n = len(matrix[0])

    i = 0
    j = n-1
    while i<m and j>=0:
        if matrix[i][j]==target:
            return True
        elif matrix[i][j]<target:
            i+=1
        elif matrix[i][j]>target:
            j-=1
    return False
```

# Coin Change

```python
def coinChange(coins, amount):
    n = len(coins)
    def dp(rem): #function with state rem
        # base case:
        if rem<0:
            return -1
        if rem == 0:
            return 0
        
        # recursion:
        min_res = float('inf')
        for coin in coins:
            res = dp(rem-coin)
            if res != -1:
                min_res = min(res+1,min_res)
        return min_res if min_res != float('inf') else -1
    
    return dp(amount)
```

```python
def diameterOfBinaryTree(root):
    diameter = 0
    def dfs(node): #returns the max height from node to deepest node
        if not node:
            return 0
        nonlocal diameter
        left = dfs(node.left)
        right = dfs(node.right)
        diameter = max(diameter, left+right)
        return max(left,right)+1
    dfs(root)
    return diameter
        
        
```

```python
# practice graph problem: Evaluate division problem
def calcEquation(equations, values, queries):
    group_Id_weight = {}  # nodeId: (groupId, weight)
    def find(nodeId):
        if nodeId not in group_Id_weight:
            group_Id_weight[nodeId] = (nodeId, 1)
        
        grouId, nodeWeight = group_Id_weight[nodeId]

        if nodeId != grouId:
            new_gr_Id, weight = find(grouId) 
            group_Id_weight[nodeId] = (new_gr_Id, weight*nodeWeight)
        return group_Id_weight[nodeId]
    
    def union(dividend, divisor, val):
        div_id, div_w = find(dividend)
        divsor_id, divisor_w = find(divisor)

        if div_id != divsor_id:
            group_Id_weight[div_id] = (divsor_id, divisor_w*val/div_w)
    
    for (dividend, divisor), val in zip(equations, values):
        union(dividend, divisor, val)
    res = []
    for (q1, q2) in queries:
        if q1 not in group_Id_weight or q2 not in group_Id_weight:
           res.append(-1.0) 
        else:

            q1_id, q1_w = find(q1)
            q2_id, q2_w = find(q2)

            if q1_id != q2_id:
                res.append(-1)
            else:
                res.append(q1_w/q2_w)
    return res
            
```

# 1059. All Paths from Source Lead to Destination

```python
from collections import defaultdict
def leadsToDestination(n, edges, source, destination):

    # first a func to make the graph from edges: return graph
    # recursive function: input: node, graph, destination, states and returns True if all the neighbors checked and none where False or short circuit to False if any False
    # Managing the states of nodes as recursion done: None, initial color, then Gray when processing a node, and then Black when processing is done


    def recur_toDest(graph, node, destination, states):

        if states[node]:  # checking if the node is in processing state, indicating that there is a cycle 
            return states[node] == 'B'
        if len(graph[node])==0: # if there is no neighbor, then it is a leaf, check if it is destination, otherwise return False
            return node == destination
        
        states[node] = 'G'
        for nei in graph[node]:
            if not recur_toDest(graph, nei, destination, states):
                return False
        
        states[node] = 'B'
        return True


    def buildGraph(n, edges):
        graph = [[] for _ in range(n)]
        for i,j in edges:
            graph[i].append(j)        
        return graph
    

    graph = buildGraph(n, edges)
    return recur_toDest(graph, source, destination, [None]*n)
```

```python
# using DFS:
def validPath(n, edges, source, destination):

    adjList = { i: [] for i in range(n)}
    

    for i,j in edges:
        adjList[i].append(j)
    stack = [source]
    seen = set()
    while stack:
        node = stack.pop()
        if node == destination:
            return True
        if node not in seen:
            seen.add(node)
            stack.extend(adjList[node])
    
    return False


```

# Vowel Spellchecker 
practice to use setdefault() and map() in the solution

```python
from typing import List

def spellchecker(wordlist: List[str], queries: List[str]) -> List[str]:

    def devowel(word):
        return "".join(['*' if c in 'aioue' else c for c in word ])
    
    words_perfect = set(wordlist)
    words_cap = {}
    words_vow = {}

    for word in wordlist:
        word_lower = word.lower()
        words_cap.setdefault(word_lower, word)
        words_vow.setdefault(devowel(word_lower), word)
    
    def solve(query):

        if query in words_perfect:
            return query
        
        query_L = query.lower()
        if query_L in words_cap:
            return words_cap[query_L]
        
        query_LV = devowel(query_L)
        if query_LV in words_vow:
            return words_vow[query_LV]
        return ""

    return list(map(solve, queries))
```

# 2197. Replace Non-Coprime Numbers in Array

```python
def replaceNonCoprimes(nums):
    
    def cal_gcd(a,b):
        while b!=0:
            a,b = b,a%b
        return a
    def cal_lcm(a,b):
        return a//cal_gcd(a,b)*b
    def isNonCoprime(a,b):
        return cal_gcd(a,b)>1
    stack = [nums[0]]
    n = len(nums)
    for i in range(1, n):
        stack.append(nums[i])
        while len(stack)>1 and isNonCoprime(stack[-1],stack[-2]):
            a = stack.pop()
            b = stack.pop()
            stack.append(cal_lcm(a,b))
    
    return stack
```

```python
nums = [6,4,3,2,7,6,2]
replaceNonCoprimes(nums)
```

```python
# using built-in gcd:
from math import gcd
def replaceNonCoprimes(nums):
    stack = []
    for num in nums:
        stack.append(num)
        while len(stack)>1 and (g:=gcd(stack[-1],stack[-2]))>1:
            a = stack.pop()
            b = stack.pop()
            stack.append(a//g*b)
    return stack
```

# 0/1 Knapsack Problem

Given n items where each item has some weight and profit associated with it and also given a bag with capacity W, [i.e., the bag can hold at most W weight in it]. The task is to put the items into the bag such that the sum of profits associated with them is the maximum possible. 

Note: The constraint here is we can either put an item completely into the bag or cannot put it at all [It is not possible to put a part of an item into the bag].

Input:  W = 4, profit[] = [1, 2, 3], weight[] = [4, 5, 1]
Output: 3
Explanation: There are two items which have weight less than or equal to 4. If we select the item with weight 4, the possible profit is 1. And if we select the item with weight 1, the possible profit is 3. So the maximum possible profit is 3. Note that we cannot put both the items with weight 4 and 1 together as the capacity of the bag is 4.

Input: W = 3, profit[] = [1, 2, 3], weight[] = [4, 5, 6]
Output: 0

```python
from functools import lru_cache
def knapsackRec(W, val, wt, n):
    @lru_cache(None)
    def dp(i,rem):
        if i==n or rem==0:
            return 0
        if wt[i]>rem:
            return dp(i+1,rem)
        else:
            return max(val[i]+dp(i+1,rem-wt[i]), dp(i+1,rem))
    return dp(0,W)
```

# Group Anagram: (two solutions)

```python
# using hashmap and sorting: complexity: O(n.mlogm) where n is the number of strings and m is the max length of str in strings. space: O(n)
from collections import defaultdict
def groupAnagrams(strs):

    seen = defaultdict(list)

    for st in strs:
        key = ''.join(sorted(st))
        seen[key]+=[st]
    
    return list(seen.values())
```

```python
class UnionFind:
    def __init__(self):
        self.parent = {}
        self.size = {}

    def add(self, x):
        if x not in self.parent:
            self.parent[x] = x
            self.size[x] = 1

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        if x not in self.parent or y not in self.parent:
            return
        rootX, rootY = self.find(x), self.find(y)
        if rootX == rootY:
            return
        # union by size
        if self.size[rootX] < self.size[rootY]:
            rootX, rootY = rootY, rootX
        self.parent[rootY] = rootX
        self.size[rootX] += self.size[rootY]

def longestConsecutive(nums):
    uf = UnionFind()
    for num in nums:
        uf.add(num)
    for num in nums:
        if num + 1 in uf.parent:
            uf.union(num, num + 1)

    return max(uf.size.values(), default=0)


# Example
print(longestConsecutive([100,4,200,1,3,2]))   # Output: 4
print(longestConsecutive([0,3,7,2,5,8,4,6,0,1]))  # Output: 9
print(longestConsecutive([1,0,1,2]))  # Output: 3
```

```python
def findMinDifference(timePoints):
    minutes = [False]*24*60
    for time in timePoints:
        minute = int(time.split(":")[0])*60 + int(time.split(":")[1])
        if minutes[minute]:
            return 0
        minutes[minute] = True
        print(minute)
    

    first_idx = float('inf')
    last_idx = float('inf')
    prev_idx = float('inf')
    ans = float('inf')

    for i in range(24*60):
        if minutes[i]:
            if prev_idx != float('inf'):
                ans = min(ans, i-prev_idx)
            prev_idx = i
            if first_idx == float('inf'):
                first_idx = i
            last_idx = i
    return min(ans, 24*60-last_idx+first_idx)
```

```python
timePoints = ["13:24","13:29"]
findMinDifference(timePoints)
```

```python
timePoints = ["23:59","00:00"]
findMinDifference(timePoints)
```

```python
timePoints = ["00:00","23:59","00:00"]
findMinDifference(timePoints)
```

# 3071. Minimum Operations to Write the Letter Y on a Grid

```python
def minimumOperationsToWriteY(grid):
    n = len(grid)
    count_Y = [0]*3
    count_nonY = [0]*3
    for i in range(n):
        for j in range(n):
            if (i == j or i == n-1-j) and i<=n//2:
                count_Y[grid[i][j]]+=1
            elif i>n//2 and j == n//2:
                count_Y[grid[i][j]]+=1
            else:
                count_nonY[grid[i][j]]+=1
    total_Y = sum(count_Y)
    total_nonY = sum(count_nonY)

    ans = float('inf')
    for y_val in range(3):
        for non_y_val in range(3):
            if y_val == non_y_val:
                continue
            ops = (total_Y - count_Y[y_val]) + (total_nonY - count_nonY[non_y_val])
            ans = min(ops, ans)
    return ans
```

```python
grid = [[1,2,2],[1,1,0],[0,1,0]]
minimumOperationsToWriteY(grid)
```

```python
grid = [[0,1,0,1,0],[2,1,0,1,2],[2,2,2,0,1],[2,2,2,2,2],[2,1,2,2,2]]
minimumOperationsToWriteY(grid)
```
