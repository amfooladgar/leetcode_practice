# Two Sum - Leetcode 1 - HashMap - Python

```python
from collections import defaultdict
def twoSum(nums, target):
    seen = defaultdict(int)
    n = len(nums)
    for i in range(n):
        if target - nums[i] in seen:
            return [i,seen[target - nums[i]]]
        seen[nums[i]] = i
    return
    
```

```python
nums, target = [2,7,11,15], 9
twoSum(nums, target)
```

```python
nums, target = [3,2,4], 6
twoSum(nums, target)
```

```python
nums, target = [3,3], 6
twoSum(nums, target)
```

# Sliding Window: Best Time to Buy and Sell Stock - Leetcode 121 - Python

```python
import math
def maxProfit(prices):
    buy, sell = 0,1
    n = len(prices)
    bestRes = 0
    while sell<n:
        if prices[sell]<prices[buy]:
            buy = sell
        else:
            profit = prices[sell]-prices[buy]
            bestRes = max(profit, bestRes)
        sell+=1
    return bestRes
```

```python
prices = [7,1,5,3,6,4]
maxProfit(prices)
```

```python
prices = [7,6,4,3,1]
maxProfit(prices)
```

# Contains Duplicate - Leetcode 217 - Python

```python
# either space optimized solution (O(1) using sorting) or time-optimized solution using hashset
def containsDuplicate(nums):
    n = len(nums)
    hashSet = set()
    for num in nums:
        if num in hashSet:
            return True
        hashSet.add(num)
    return False

```

```python
nums = [1,2,3,1]
containsDuplicate(nums)
```

```python
nums = [1,2,3,4]
containsDuplicate(nums)
```

# Product of Array Except Self - Leetcode 238 - Python

```python
# solution 1: using prefix and postfix arrays O(N) space. This can be later optimized:

def productExceptSelf(nums):
    n = len(nums)
    prefix = [1]*n
    postfix = [1]*n
    prefix[0] = nums[0]
    postfix[n-1] = nums[n-1]
    for i in range(1,n):
        prefix[i] = nums[i]*prefix[i-1]
    for i in range(n-2,-1,-1):
        postfix[i] = nums[i]*postfix[i+1]
    
    res = [0]*n
    for i in range(n):
        if i == 0:
            res[i] = postfix[i+1]
        elif i == n-1:
            res[i] = prefix[i-1]
        else:
            res[i] = prefix[i-1]*postfix[i+1]
    return res
```

```python
nums = [1,2,3,4]
productExceptSelf(nums)
```

```python
nums = [-1,1,0,-3,3]
productExceptSelf(nums)
```

```python
# Now this can be space-optimized by doing prefix and postfix on the fly
def productExceptSelf(nums):
    n = len(nums)
    res = [1]*n
    prefix = 1
    for i in range(n):
        res[i] = prefix
        prefix *= nums[i]
    
    postfix = 1
    for i in range(n-1,-1,-1):
        res[i] *= postfix
        postfix *= nums[i]
    return res
```

```python
nums = [1,2,3,4]
productExceptSelf(nums)
```

```python
nums = [-1,1,0,-3,3]
productExceptSelf(nums)
```

# Maximum Subarray - Amazon Coding Interview Question - Leetcode 53 - Python

```python
def maxSubArray(nums):
    maxSub = nums[0]
    currSum = 0
    for n in nums:
        if currSum<0:
            currSum = 0
        currSum+=n
        maxSub = max(maxSub, currSum)
    return maxSub
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
maxSubArray(nums)
```

```python
nums = [5,4,-1,7,8]
maxSubArray(nums)
```

```python
nums = [1]
maxSubArray(nums)
```

# Maximum Product Subarray - Dynamic Programming - Leetcode 152

```python
def maxProduct(nums):

    curr_max, curr_min = 1, 1
    n = len(nums)
    maxProd = max(nums)
    for i in range(n):
        if nums[i] == 0:
            curr_max, curr_min = 1, 1
        tmp = curr_max
        curr_max = max(nums[i]*curr_min, nums[i]*curr_max, nums[i])
        curr_min = min(nums[i]*curr_min, nums[i]*tmp, nums[i])
        
        maxProd = max(curr_max, maxProd)
    return maxProd

```

```python
nums = [2,3,-2,4]
maxProduct(nums)
```

# Find Minimum in Rotated Sorted Array - Binary Search - Leetcode 153 - Python

```python
import math
def findMin(nums):
    l, r = 0, len(nums) - 1
    res = nums[0]
    while l<=r:
        if nums[l]<nums[r]:
            return min(res, nums[l])
        
        mid = (l+r)//2
        res = min(res, nums[mid])
        if nums[mid]>=nums[l]:
            l = mid + 1
        else:
            r = mid - 1
    return res
```

```python
nums = [3,4,5,1,2]
findMin(nums)
```

```python
nums = [4,5,6,7,0,1,2]
findMin(nums)
```

```python
nums = [11,13,15,17]
findMin(nums)
```

# Search in rotated sorted array - Leetcode 33 - Python

```python
def search(nums, target):
    l, r = 0, len(nums) - 1

    while l<=r:
        mid = (l+r)//2

        if nums[mid] == target:
            return mid
        
        if nums[mid]>=nums[l]:  #search on the left-sorted portion
            if target>nums[mid] or target<nums[l]:
                l=mid+1
            else:
                r = mid - 1
        else:  # search on the right-sorted portion
            if target<nums[mid] or target > nums[r]:
                r = mid -1
            else:
                l = mid + 1
    return -1
```

```python
nums, target = [4,5,6,7,0,1,2],  0
search(nums, target)
```
