# 767 Reorganize String
Given a string s, rearrange the characters of s so that any two adjacent characters are not the same.

Return any possible rearrangement of s or return "" if not possible.

 

Example 1:
```
Input: s = "aab"
Output: "aba"
```
Example 2:
```
Input: s = "aaab"
Output: ""
```

Constraints:

- 1 <= s.length <= 500
- s consists of lowercase English letters.

Hint:
- Alternate placing the most common letters.

````python
import heapq

def reorganizeString(s):
    # Step 1: Count frequency of each character
    freq_map = [0] * 26
    for ch in s:
        freq_map[ord(ch) - ord('a')] += 1

    max_count = max(freq_map)
    n = len(s)
    
    # Step 2: Early check â€“ if the most frequent char is too frequent, return ""
    if max_count > (n + 1) // 2:
        return ""

    # Step 3: Create a max heap of (-freq, char)
    max_heap = [(-freq_map[i], chr(i + ord('a'))) for i in range(26) if freq_map[i] > 0]
    heapq.heapify(max_heap)

    res = []
    prev_count, prev_char = 0, ''

    # Step 4: Build the result string
    while max_heap:
        count, char = heapq.heappop(max_heap)
        res.append(char)

        # If there's a previous character held back, push it back into the heap
        if prev_count < 0:
            heapq.heappush(max_heap, (prev_count, prev_char))

        # Hold the current char out for one round (decrement its count)
        prev_count, prev_char = count + 1, char

    return ''.join(res)
````

````python
s = "aaabccs"
reorganizeString(s)
````

````python
def removeDuplicates(nums):
    n = len(nums)
    if n == 0:
        return 0
    k, r = 1, 1
    while r < n:
        if nums[r] != nums[k-1]:
            nums[k] = nums[r]
            k += 1
        r += 1
    return k
````

````python
# Leetcode solution
class Solution:
    def removeDuplicates(self, nums) -> int:
        size = len(nums)
        insertIndex = 1
        for i in range(1, size):
            # Found unique element
            if nums[i - 1] != nums[i]:
                # Updating insertIndex in our main array
                nums[insertIndex] = nums[i]
                # Incrementing insertIndex count by 1
                insertIndex = insertIndex + 1
        return insertIndex
````

Two Sum Variant
Intermediate
Given an array of integers and a target sum, find all unique pairs that sum to the target. Return pairs as tuples sorted by first element.

Example:
Input: [1, 2, 3, 4, 5], target = 6
Output: [(1, 5), (2, 4)]
Test Cases:
Input: ([1, 2, 3, 4, 5], 6)
Expected: [(1, 5), (2, 4)]
Input: ([1, 1, 2, 3], 4)
Expected: [(1, 3)]

````python
# Ali's solution: 
def twoSum(nums, target):
    nums.sort()
    low = 0
    high = len(nums)-1
    ans = []
    while low<high:
        if nums[high]<target - nums[low]:
            low+=1
        elif nums[high]>target - nums[low]:
            high-=1
        else:
            ans.append((nums[low],nums[high]))
            low+=1
    return ans
````

````python
# Fixed by chatgpt as it was missing duplicates
def twoSum(nums, target):
    nums.sort()
    low, high = 0, len(nums) - 1
    ans = []

    while low < high:
        s = nums[low] + nums[high]
        if s < target:
            low += 1
        elif s > target:
            high -= 1
        else:
            ans.append((nums[low], nums[high]))
            low += 1
            high -= 1
            # Skip duplicates
            while low < high and nums[low] == nums[low - 1]:
                low += 1
            while low < high and nums[high] == nums[high + 1]:
                high -= 1
    return ans
````
